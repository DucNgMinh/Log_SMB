```python
excel_data = pd.read_excel(r'C:\Users\Admin\PycharmProjects\Log SmartBanking\Data\Book1.xlsx')

# Convert date columns to datetime for analysis
excel_data['Ngày yêu cầu'] = pd.to_datetime(excel_data['Ngày yêu cầu'])
excel_data['Ngày phản hồi'] = pd.to_datetime(excel_data['Ngày phản hồi'])

# Analyzing Request Types and Frequency
request_analysis = excel_data['Request'].value_counts().head(20)

# Success vs Failure Rates
status_analysis = excel_data['Trạng thái'].value_counts()

# Temporal Analysis
# Extract hour from request time
excel_data['Request Hour'] = excel_data['Ngày yêu cầu'].dt.hour
hourly_analysis = excel_data['Request Hour'].value_counts().sort_index()

# Response Times Analysis
excel_data['Response Time'] = (excel_data['Ngày yêu cầu'] - excel_data['Ngày phản hồi']).dt.total_seconds()
response_time_analysis = excel_data['Response Time'].describe()

# Plotting
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Request Analysis Plot
axs[0, 0].bar(request_analysis.index, request_analysis.values, color='skyblue')
axs[0, 0].set_title('Request Types Frequency')
axs[0, 0].set_xticklabels(request_analysis.index, rotation=45, ha="right")
axs[0, 0].set_ylabel('Frequency')

# Status Analysis Plot
axs[0, 1].pie(status_analysis, labels=status_analysis.index, autopct='%1.1f%%', startangle=140)
axs[0, 1].set_title('Success vs Failure Rates')

# Hourly Analysis Plot
axs[1, 0].plot(hourly_analysis.index, hourly_analysis.values, marker='o', color='orange')
axs[1, 0].set_title('Hourly Request Distribution')
axs[1, 0].set_xlabel('Hour of Day')
axs[1, 0].set_ylabel('Number of Requests')

# Response Time Analysis Plot
axs[1, 1].boxplot(excel_data['Response Time'].dropna())
axs[1, 1].set_title('Response Time Analysis (Seconds)')
axs[1, 1].set_ylabel('Response Time (Seconds)')

plt.tight_layout()
plt.show()

# Print summary statistics for response time
print(response_time_analysis)
```

```python
status_by_request = excel_data.groupby(['Request'])['Trạng thái'].value_counts().unstack().fillna(0)

# Calculate Failure Rate and filter requests with > 20% failure rate
failure_rate = status_by_request.apply(lambda x: x.get('False', 0) / x.sum(), axis=1)
requests_with_high_failure = status_by_request[failure_rate > 0.15]

# Plotting
fig = plt.figure(figsize=(40, 20))
for i, (request, data) in enumerate(requests_with_high_failure.iterrows(), 1):
    ax = fig.add_subplot(len(requests_with_high_failure), 1, i, aspect='equal')
    ax.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140)
    ax.set_title(request)

plt.tight_layout()
plt.show()
```

```python

# Sort the data by 'IMei', 'Request', and then by 'Ngày yêu cầu' (date of request) for sequential analysis
excel_data.sort_values(by=['IMei', 'Request', 'Ngày yêu cầu'], inplace=True)

# Function to detect if there are at least 3 consecutive False statuses
def detect_three_consecutive_false(group):
    # Create a boolean array where 'False' statuses are 1 and others are 0
    status_array = group['Trạng thái'].map({'False': 1}).fillna(0).astype(int)
    # Use a rolling window of 3 to check for three consecutive 1's (False statuses)
    group['Three Consecutive False'] = status_array.rolling(window=3).sum() == 3
    return group

# Apply the function to each group of IMEIs
consecutive_false_analysis = excel_data.groupby('IMei').apply(detect_three_consecutive_false)

# Filter the data to find where 'Three Consecutive False' is True
three_consecutive_false_imeis = consecutive_false_analysis[consecutive_false_analysis['Three Consecutive False']]['IMei'].unique()


```
